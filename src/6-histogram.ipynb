### Exercise 6: Histogram Customization

```python
# TASK: Create a histogram for the following data: data = [1, 2, 2, 3, 3, 3, 4, 4, 4, 4, 5].
# Customize the histogram with a title, labels for the x-axis, and a specific color for the bars.
import matplotlib.pyplot as plt
import numpy as np

def validate_histogram_data(data):
    """Validate that data is a NumPy array or a list containing numeric values."""
    if not isinstance(data, (np.ndarray, list)):
        raise TypeError("Histogram: Data should be a NumPy array or a list.")
    if isinstance(data, list) and not all(isinstance(i, (int, float)) for i in data):
        raise TypeError("Histogram: All elements in data must be integers or floats.")
    if len(data) == 0:
        raise ValueError("Histogram: Data should not be empty.")

data = np.random.randn(1000)

try:
    validate_histogram_data(data)

    plt.hist(data, bins=30, edgecolor='black')

    plt.title('Histogram Example')
    plt.xlabel('Value')
    plt.ylabel('Frequency')

    plt.show()
except (TypeError, ValueError) as e:
    print(f"Error: {e}")
```
