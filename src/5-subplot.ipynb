### Exercise 5: Subplot Layout

```python
# TASK: Create a 2x2 subplot layout.
# Plot a line chart in the first subplot, a bar chart in the second, a scatter plot in the third, and a pie chart in the fourth.
import matplotlib.pyplot as plt

def validate_numeric_data(x, y, name="Data"):
    """Validate that x and y are lists of the same length containing numeric values."""
    if not isinstance(x, list) or not isinstance(y, list):
        raise TypeError(f"{name}: Both x and y should be lists.")
    if len(x) != len(y):
        raise ValueError(f"{name}: x and y must have the same length.")
    if not all(isinstance(i, (int, float)) for i in x + y):
        raise TypeError(f"{name}: All elements in x and y must be integers or floats.")

def validate_bar_data(categories, values):
    """Validate that categories and values are lists of the same length."""
    if not isinstance(categories, list) or not isinstance(values, list):
        raise TypeError("Bar Plot: categories and values should be lists.")
    if len(categories) != len(values):
        raise ValueError("Bar Plot: categories and values must have the same length.")
    if not all(isinstance(cat, str) for cat in categories):
        raise TypeError("Bar Plot: All categories must be strings.")
    if not all(isinstance(val, (int, float)) and val >= 0 for val in values):
        raise TypeError("Bar Plot: All values must be non-negative integers or floats.")

def validate_pie_data(sizes, labels):
    """Validate that sizes and labels are lists of the same length and that sizes sum > 0."""
    if not isinstance(sizes, list) or not isinstance(labels, list):
        raise TypeError("Pie Chart: sizes and labels should be lists.")
    if len(sizes) != len(labels):
        raise ValueError("Pie Chart: sizes and labels must have the same length.")
    if not all(isinstance(size, (int, float)) and size >= 0 for size in sizes):
        raise TypeError("Pie Chart: All sizes must be non-negative integers or floats.")
    if not all(isinstance(label, str) for label in labels):
        raise TypeError("Pie Chart: All labels must be strings.")
    if sum(sizes) == 0:
        raise ValueError("Pie Chart: The sum of sizes must be greater than zero.")


x = [1, 2, 3, 4, 5]
y1 = [1, 4, 9, 16, 25]
y2 = [5, 4, 3, 2, 1]
bar_categories = ['A', 'B', 'C']
bar_values = [3, 7, 2]
pie_sizes = [15, 30, 45, 10]
pie_labels = ['A', 'B', 'C', 'D']

try:

    validate_numeric_data(x, y1, "Line Plot")
    validate_numeric_data(x, y2, "Scatter Plot")
    validate_bar_data(bar_categories, bar_values)
    validate_pie_data(pie_sizes, pie_labels)

 
    fig, axs = plt.subplots(2, 2)

    axs[0, 0].plot(x, y1)
    axs[0, 0].set_title('Line Plot')

    axs[0, 1].scatter(x, y2)
    axs[0, 1].set_title('Scatter Plot')

    axs[1, 0].bar(bar_categories, bar_values)
    axs[1, 0].set_title('Bar Plot')

    axs[1, 1].pie(pie_sizes, labels=pie_labels, autopct='%1.1f%%')
    axs[1, 1].set_title('Pie Chart')

    plt.tight_layout()
    plt.show()

except (TypeError, ValueError) as e:
    print(f"Error: {e}")
```
